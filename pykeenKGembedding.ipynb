## PyKEEN Knowledge Graph Embeddings on the created movie graph

```python
import torch
from pykeen.pipeline import pipeline
from pykeen.triples import TriplesFactory
import pandas as pd

# Load your triples file
file_path = 'data/kg/movie_kg_triples.tsv'
tf = TriplesFactory.from_path(file_path)

# Split into training and testing sets (80/20 split)
training_tf, testing_tf = tf.split([0.8, 0.2])

# Train the model
result = pipeline(
    training=training_tf,
    testing=testing_tf,
    model='TransE',
    training_kwargs=dict(num_epochs=100),
    random_seed=42,
    device='cuda' if torch.cuda.is_available() else 'cpu',
)

# Extract embeddings
entity_rep = result.model.entity_representations[0]
relation_rep = result.model.relation_representations[0]

entity_embeddings = entity_rep(torch.arange(tf.num_entities)).detach().numpy()
relation_embeddings = relation_rep(torch.arange(tf.num_relations)).detach().numpy()

# Save to CSV
entity_df = pd.DataFrame(entity_embeddings, index=tf.entity_labeling.label_to_id.keys())
relation_df = pd.DataFrame(relation_embeddings, index=tf.relation_labeling.label_to_id.keys())

entity_df.to_csv("data/kg/embeddings/entity_embeddings.csv")
relation_df.to_csv("data/kg/embeddings/relation_embeddings.csv")

print("âœ… Training complete. Embeddings saved to CSV.")
```

## Visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# === 1. Lade die CSV-Dateien ===
entity_embeddings = pd.read_csv("data/kg/embeddings/entity_embeddings.csv")
relation_embeddings = pd.read_csv("data/kg/embeddings/relation_embeddings.csv")

# === 2. Trenne Vektoren und Labels ===
entity_vectors = entity_embeddings.drop(columns=["Unnamed: 0"])
relation_vectors = relation_embeddings.drop(columns=["Unnamed: 0"])

entity_labels = entity_embeddings["Unnamed: 0"].astype(str).fillna("Unknown").values
relation_labels = relation_embeddings["Unnamed: 0"].astype(str).values

# === 3. Reduziere Dimensionen mit PCA ===
pca = PCA(n_components=2)
entity_2d = pca.fit_transform(entity_vectors)
relation_2d = pca.transform(relation_vectors)

# === 4. WÃ¤hle zufÃ¤llige Stichprobe von EntitÃ¤ten fÃ¼r die Darstellung ===
sample_size = 10000
sample = entity_embeddings.sample(sample_size, random_state=42)
sample_indices = sample.index
sample_entity_2d = entity_2d[sample_indices]
sample_entity_labels = entity_labels[sample_indices]

# === 5. Plotten ===
plt.figure(figsize=(12, 8))
plt.scatter(sample_entity_2d[:, 0], sample_entity_2d[:, 1], alpha=0.5, label="Entities")
plt.scatter(relation_2d[:, 0], relation_2d[:, 1], color='red', marker='x', s=100, label="Relations")

# Beschrifte die Relationen
for i, label in enumerate(relation_labels):
    plt.annotate(label, (relation_2d[i, 0], relation_2d[i, 1]), fontsize=9, color='red')

plt.title("2D PCA-Projektion der Embeddings")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
